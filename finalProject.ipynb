### Visualize the data

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
parentPath='/content/drive/MyDrive/Data Science/GANFinalProject'
directory=os.listdir(parentPath)
counter=0
fig=plt.figure(figsize=(10,10))
for subfolder in directory:
    counter= counter + 1
    currentFolder = parentPath +'/'+ subfolder+'/TrainT'+subfolder.split('Tr')[1]+'/'
    plt.subplot(3,3,counter+1)
    for i, file in enumerate(os.listdir(currentFolder)[0:1]):
        completeImagePath = currentFolder+ "/" + file
        img=mpimg.imread(completeImagePath)
        plt.imshow(img, cmap='gray')
        plt.title(subfolder)
        plt.axis("off")
```
